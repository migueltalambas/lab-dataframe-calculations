## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python

# Sub 1

# groupby to aggregate and then sum the grouped values
unique = orders.groupby(['CustomerID', 'Country']).sum()

# Sub 2

## Identify which quartile represents which variable
VIP = unique['amount_spent'].quantile(0.95)

Preferred_Customers = unique['amount_spent'].quantile([0.95, 0.75])

print(VIP)
print(Preferred_Customers)

# Sub 3

# used the loc to find only the ones that meet the condition and update the 'Costumer Label' column
unique.loc[unique['amount_spent'] > unique['amount_spent'].quantile(0.75), 'Costumer Label'] = 'PC'
unique.loc[unique['amount_spent'] > unique['amount_spent'].quantile(0.95), 'Costumer Label'] = 'VIP'

unique.head(40)
```

Now we'll leave it to you to solve Q2 & Q3, which you can leverage from your solution for Q1:

## Q2: How to identify which country has the most VIP Customers?

```python
# your code here
condition = unique['Costumer Label'] == 'VIP'
x = unique[condition]

# Goruped by country and counted the number of rows
# Note that each row represents one VIP customer so the number of rows is the same as the number of VIP's
organized = x.groupby(['Country']).count()

# Sorted to make it ordered
sort = organized.sort_values(by = 'Costumer Label', ascending = False)

# got the top country!
top_country = sort.head(1)
top_country
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# your code here
# Since the 'Costumer Label' either has VIP or PC I organized them by country and then did the count of the rows.
# Some rows will have 'NaN' but with the .count in the 'Costumer Label' will only count the ones that has a value -> VIP or PC
organized = unique.groupby(['Country']).count()

# Sorted by 'Costumer Label'
sort = organized.sort_values(by = 'Costumer Label', ascending = False)

# Get the top country
sort.head(1)
```
